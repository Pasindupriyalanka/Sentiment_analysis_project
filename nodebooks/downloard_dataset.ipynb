```python
import json
import os
from pathlib import Path

# Get the current working directory and find the project root
current_dir = Path.cwd()
kaggle_json_path = None

# Try different possible paths
possible_paths = [
    current_dir / 'artifacts' / 'kaggle.json',  # If running from project root
    current_dir.parent / 'artifacts' / 'kaggle.json',  # If running from nodebooks
    Path('../artifacts/kaggle.json'),  # Relative path
    Path('artifacts/kaggle.json'),  # Direct path
]

for path in possible_paths:
    if path.exists():
        kaggle_json_path = path
        break

if kaggle_json_path is None:
    raise FileNotFoundError(f"Could not find kaggle.json. Checked: {[str(p) for p in possible_paths]}")

with open(kaggle_json_path, 'r') as file:
    data = json.load(file)
    
print("Kaggle credentials loaded successfully!")
print(f"Username: {data.get('username', 'N/A')}")
```
